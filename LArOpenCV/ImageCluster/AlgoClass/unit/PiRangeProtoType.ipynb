```python
class opening:
    
    def __init__(self):
        self.angle=0
        self.hi=0
        self.lo=0
        self.angle_hi = self.angle_lo = 0
        
    def set(self,angle,lo,hi):
        self.angle = float(angle)
        self.lo    = float(lo)
        self.hi    = float(hi)
        if self.lo < 0 or self.lo > 180:
            raise Exception
        if self.hi < 0 or self.hi > 180:
            raise Exception
        if self.angle > 360: self.angle -= 360.
        if self.angle < 0:   self.angle += 360.
            
        self.angle_hi = self.angle + self.hi
        self.angle_lo = self.angle - self.lo
        
        if self.angle_hi > 360: self.angle_hi -= 360.
        if self.angle_hi < 0:   self.angle_hi += 360.
        if self.angle_lo > 360: self.angle_lo -= 360.
        if self.angle_lo < 0:   self.angle_lo += 360.
            
    def inside(self,a):
        # check if it is closer to lo or hi
        dangle_hi = self.angle_hi - a
        if dangle_hi < 0   : dangle_hi *= -1.
        if dangle_hi > 180.: dangle_hi = 360. - dangle_hi
        
        dangle_lo = self.angle_lo - a
        if dangle_lo < 0   : dangle_lo *= -1.
        if dangle_lo > 180.: dangle_lo = 360. - dangle_lo
            
        print 'hi', self.angle_hi, a, dangle_hi
        print 'lo', self.angle_lo, a, dangle_lo
        if dangle_hi > dangle_lo:
            print 'closer to lo',
            #rotate to make angle_lo @0
            angle_rotate = 360 - self.angle_lo
            a_rotated = a + angle_rotate
            if a_rotated > 360: a_rotated -= 360
            if a_rotated >180: print 'outside'
            else: print 'inside'
        else:
            print 'closer to hi',
            # rotate to make angle_hi @0
            angle_rotate = 360 - self.angle_hi
            a_rotated = a + angle_rotate
            if a_rotated > 360: a_rotated -= 360
            if a_rotated > 180: print 'inside'
            else: print 'outside'
        print
        
angle = random.random() * 360
lo = random.random() * 180
hi = random.random() * 180
c=geo2d.Circle(DTYPE)()
c.center.x, c.center.y = (0., 0.)
c.radius = 2.

print 'angle',angle,'lo',lo,'hi',hi
print

k=opening()
k.set(angle,lo,hi)

for x in xrange(36):
    a = x*10
    k.inside(a)
    
    # Prepare canvas
    fig,ax = plt.subplots(figsize=(4,4))

    # Define circle
    cplt = plt.Circle((c.center.x,c.center.y),c.radius,color='lightgray')
    ax.add_artist(cplt)
    # put points
    plt.plot([0,2*np.cos(2*3.14*k.angle/360.)],[0,2*np.sin(2*3.14*k.angle/360.)],color='black',marker='o')
    plt.plot([0,2*np.cos(2*3.14*k.angle_lo/360.)],[0,2*np.sin(2*3.14*k.angle_lo/360.)],color='b',marker='o')
    plt.plot([0,2*np.cos(2*3.14*k.angle_hi/360.)],[0,2*np.sin(2*3.14*k.angle_hi/360.)],color='r',marker='o')
    plt.plot([np.cos(2*3.14*a/360.)],[np.sin(2*3.14*a/360.)],color='black',marker='o',markersize=10)
    plt.xlim(-2,2)
    plt.ylim(-2,2)
    plt.show()
```
