```python
import ROOT
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as ptch
from ROOT import larocv
import numpy as np
%matplotlib inline
```

```python
pirange=larocv.PiRange()
pirange.set_verbosity(0)
```

```python
for aa in xrange(0,360,36):
    pirange.SetAngle(aa,45,90)
    print "Angle:   ",pirange.Angle()
    print "AngleLo: ",pirange.AngleLo()
    print "AngleHi:  ",pirange.AngleHi()

    fig,ax = plt.subplots(figsize=(5,5))
    circ=ptch.Circle((0,0),radius=3,fc='lightgray')
    ax.add_patch(circ)
    ax.set_xlim(-3.1,3.1)
    ax.set_ylim(-3.1,3.1)
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    a=50.0
    plt.plot([0,3*np.cos(2*3.14*pirange.Angle()/360.)],
             [0,3*np.sin(2*3.14*pirange.Angle()/360.)],
             color='black',marker='o')
    plt.plot([0,3*np.cos(2*3.14*pirange.AngleLo()/360.)],
             [0,3*np.sin(2*3.14*pirange.AngleLo()/360.)],
             color='blue',marker='o')
    plt.plot([0,3*np.cos(2*3.14*pirange.AngleHi()/360.)],
             [0,3*np.sin(2*3.14*pirange.AngleHi()/360.)],
             color='red',marker='o')
    for a in xrange(-360*2,360*2,5):
        if a < -360:
            r=0.5
        elif a < 0:
            r=1.0
        elif a < 360:
            r=1.5
        else:
            r=2.0

        if pirange.Inside(a)==True:
            color='green'
        else:
            color='red'

        plt.plot([r*np.cos(2*3.14*a/360.)],
                 [r*np.sin(2*3.14*a/360.)],
                 color=color,marker='o',markersize=5)
    plt.grid()
    print
    print "From inner to outer..."
    print "0) -720 -> -360"
    print "1) -360 ->    0"
    print "2)    0 ->  360"
    print "3)  360 ->  720"
    plt.show()
```
